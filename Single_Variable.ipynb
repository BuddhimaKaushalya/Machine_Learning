```python
x = np.array(data.videos.values)
x
```

```python
y = np.array(data.views.values)
y
```

```python
model = LinearRegression()
model.fit(x.reshape((-1,1)),y)
```

```python
new_x = np.array([45]).reshape((-1,1))
new_x
```

```python
pred = model.predict(new_x)
pred
```
