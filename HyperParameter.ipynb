```python
import pandas as pd
import numpy as np
```

```python
data = pd.read_csv("iris.csv")
x = data.drop(['Species','Id'] , axis =1)
y = data['Species']
```

```python
x.sample(5)
```

```python
y.sample(5)
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.3)
```

```python
from sklearn.svm import SVC
model1 = SVC()
```

```python
model1.fit(x_train,y_train)
```

```python
model1.score(x_test,y_test)
```

```python
from sklearn.svm import SVC
model = SVC(C=1,kernel='linear')
```

```python
model.fit(x_train,y_train)
```

```python
model.score(x_test,y_test)
```

# GridSearchCV

```python
from IPython.display import Image
```

```python
Image(filename= 'Capture1.png')
```

```python
from sklearn.model_selection import GridSearchCV
```

```python
param_grid ={ 'C' : [0.1,1,10],
             'kernel': ['rbf','linear', 'poly']}
    
```

```python
grid_search = GridSearchCV(estimator=model1, param_grid=param_grid)
grid_search.fit(x_train,y_train)
```

```python
grid_search.best_params_
```

```python
grid_search.score(x_test,y_test)
```

# RandomizedSearchCV

```python
Image(filename= 'Capture2.png')
```

```python
from sklearn.model_selection import RandomizedSearchCV
```

```python
param_dist = { 'C' :[0.1,1,10],
              'kernel': ['rbf','linear','poly']}
```

```python
randomized_search = RandomizedSearchCV(estimator = model1, param_distributions = param_dist,n_iter=8)
randomized_search.fit(x_train,y_train)
```

```python
randomized_search.best_params_
```

```python
randomized_search.score(x_test,y_test)
```
