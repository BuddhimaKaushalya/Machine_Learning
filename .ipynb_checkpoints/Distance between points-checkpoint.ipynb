# Euclidean Distance


```python
A = (2,3)
B = (5,4)
from scipy.spatial import distance
```

```python
# Euclidean Distance

d1 = distance.euclidean(A,B)
print(d1)
```

```python
C = (1,0,2,5,3)
D = (2,1,0,3,-1)
```

```python
d2 = distance.euclidean(C,D)
print(d2)
```
